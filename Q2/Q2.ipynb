1. it is derived from matrix inverse.

$$
x[n] = \frac{1}{N} \sum_{k=0}^{N-1} X[k] e^{j\frac{ 2\pi}{N}k n}
$$

```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
def plot(xlabel: str, ylabel: str, title: str, x, y):
    plt.figure(figsize=(30, 10))
    plt.plot(x, y)
    plt.xlabel(xlabel)
    plt.ylabel(ylabel)
    plt.title(title)
    plt.show()
```

```python
def fft(x: np.ndarray):
    N = x.shape[0]
    w = np.exp(-1.0j * 2 * np.pi / N)
    F = np.zeros((N, N), dtype=complex)
    for i, row in enumerate(F):
        for j, entry in enumerate(row):
            F[i, j] = w ** (i * j)
    return F @ x
```

3.

```python
def ifft(x: np.ndarray):
    N = x.shape[0]
    w = np.exp(1.0j * 2 * np.pi / N)
    F = np.zeros((N, N), dtype=complex)
    for i, row in enumerate(F):
        for j, entry in enumerate(row):
            F[i, j] = w ** (i * j)
    F /= N
    return F @ x
```

```python
n = np.linspace(-500, 500, 1000, dtype=int)
x = np.cos(np.pi * n * 0.5)
plot('','x', '', n, x)
```

6.

```python
fourier_x = fft(x)
plot('', 'X', 'fft x', n, np.abs(fourier_x))
```

7.

```python
inverse_fourier_x = ifft(fourier_x)
plot('', 'x', 'x', n, inverse_fourier_x.real)
print(f'equal with x: {(np.abs(inverse_fourier_x) == x)[0]}')
```

```python
list = [(2000, 0.25), (4000, 0.25), (2000, 0.5), (4000, 0.5), (2000, 1), (4000, 1)]
for N, T in list:
    n = np.linspace(-N/2, N/2, N, dtype=int)
    x = np.cos(np.pi * n * T)
    fourier_x = fft(x)
    plot('','fft x', f'N = {N}, T = {T}', n, np.abs(fourier_x))
```
