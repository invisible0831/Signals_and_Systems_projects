```python
def plot(xlabel: str, ylabel: str, title: str, x, y, ylim):
    plt.figure(figsize=(30, 10))
    plt.semilogx(x, y)
    plt.xlabel(xlabel)
    plt.ylabel(ylabel)
    plt.title(title)
    plt.ylim(ylim)
    plt.show()
```

$$
H(j\omega) = \frac{(j\omega)^2 + 2j\omega + 10}{(j\omega)^3 + 25(j\omega)^2 + 1.9 \times 10^6 j\omega + 10}
$$

```python
freq = np.logspace(-8, 8, 1000)
w = 2 * np.pi * freq
s = 1.0j * w
Hjw = (s**2 + 2*s + 10) / (s**3 + 25*(s**2) + 1.9 * (10**6) *s + 10)
```

```python
log_magnitude = 20 * np.log10(np.abs(Hjw))
phase = 180 / np.pi * np.arctan2(np.imag(Hjw), np.real(Hjw))
```

```python
plot('frequency(Hz)', '20log10|H(jw)|', 'without control', freq, log_magnitude, [-180, 0])
```

```python
plot('frequency(Hz)', 'phase(degree)', 'without control', freq, phase, [-90, 90])
```

```python
# with control package
Hjw = control.TransferFunction([1.0, 2.0, 10.0], [1.0, 25.0, 1.9*(10**6), 10])
diagram = control.bode_plot(Hjw, w, dB=1, Hz=1, deg=1)
```

```python
zeros = Hjw.zero()
for zero in zeros:
    print(f'zero: {zero:.3f}')
```

```python
poles = Hjw.pole()
for pole in poles:
    print(f'pole: {pole:.3f}')
```
